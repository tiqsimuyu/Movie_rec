```python
import pandas as pd

anime_data = {
    "title": [
        "Attack on Titan",
        "Death Note",
        "Jujutsu Kaisen",
        "Demon Slayer",
        "Tokyo Ghoul",
        "Fullmetal Alchemist: Brotherhood",
        "Hunter x Hunter",
        "Chainsaw Man",
        "My Hero Academia",
        "Naruto Shippuden"
    ],

    "genres": [
        "Action Drama DarkFantasy",
        "Thriller Mystery Psychological",
        "Action Supernatural DarkFantasy",
        "Action Adventure DarkFantasy",
        "Action Horror DarkFantasy",
        "Adventure Action Fantasy",
        "Adventure Action Fantasy",
        "Action Horror DarkFantasy",
        "Action Superhero Fantasy",
        "Action Adventure Ninja"
    ]
}

anime = pd.DataFrame(anime_data)
anime
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Convert genres to TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(anime['genres'])

# Similarity matrix
similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)

def recommend_anime(title):
    if title not in anime['title'].values:
        print("Anime not found, babe.")
        return

    idx = anime[anime['title'] == title].index[0]
    scores = list(enumerate(similarity[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print(f"\nAnime similar to '{title}':\n")
    for i in scores[1:4]:  # top 3
        print(anime.iloc[i[0]]['title'])
```

```python
recommend_anime("Demon Slayer")
```
