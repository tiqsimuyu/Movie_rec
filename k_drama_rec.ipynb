```python
import pandas as pd

data = {
    "title": [
        "Crash Landing on You",
        "Goblin",
        "Descendants of the Sun",
        "Vincenzo",
        "True Beauty",
        "Business Proposal",
        "It's Okay to Not Be Okay",
        "Twenty Five Twenty One",
        "My Demon",
        "Alchemy of Souls"
    ],
    "genres": [
        "Romance Comedy Melodrama",
        "Fantasy Romance Drama",
        "Romance Melodrama Action",
        "Crime Action Dark Comedy",
        "Romance Comedy School",
        "Romantic Comedy",
        "Romance Psychology Healing",
        "Romance Youth Drama",
        "Romance Fantasy Supernatural",
        "Fantasy Romance Action"
    ]
}

kdramas = pd.DataFrame(data)
kdramas
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Convert genres to TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(kdramas['genres'])

# Calculate similarity
similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)

def recommend(drama_title):
    if drama_title not in kdramas['title'].values:
        print("Drama not found, sweetheart.")
        return

    idx = kdramas[kdramas['title'] == drama_title].index[0]
    scores = list(enumerate(similarity[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print(f"\nK-dramas similar to '{drama_title}':\n")
    for s in scores[1:4]:  # top 3
        print(kdramas.iloc[s[0]]['title'])
```

```python
recommend("Goblin")
```

```python
recommend("Crash Landing on You")
```
