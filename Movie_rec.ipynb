```python
import pandas as pd

data = {
    "title": ["Interstellar", "Inception", "The Dark Knight", "Tenet", "Memento", "Dunkirk"],
    "genres": [
        "Sci-Fi Adventure Drama",
        "Sci-Fi Thriller Action",
        "Action Crime Drama",
        "Sci-Fi Action",
        "Mystery Thriller",
        "War Drama History"
    ]
}

movies = pd.DataFrame(data)
movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Convert genres to TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['genres'])

# Similarity scores
similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)

def recommend(movie_title):
    if movie_title not in movies['title'].values:
        print("Movie not found, babe.")
        return

    idx = movies[movies['title'] == movie_title].index[0]
    scores = list(enumerate(similarity[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print(f"\nMovies similar to '{movie_title}':\n")
    for s in scores[1:4]:  # top 3
        print(movies.iloc[s[0]]['title'])
```

```python
recommend("Interstellar")
```
